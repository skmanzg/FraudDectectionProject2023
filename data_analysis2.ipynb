```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## CC_NUM과 DAY컬럼으로 정렬된 새로운 훈련 데이터를 불러온다

```python
df_train_ordered = pd.read_csv("data/ordered_by_ccnum_and_day.csv", index_col = 0)
df_train_ordered.head()
```

## 하나의 CC_NUM에 대한 AMT의 시간순 변화를 그려주는 그래프이다.
### 그래프를 보면, FRAUD(노란 점)인 경우에 AMT가 심하게 요동치는 것을 볼 수 있다.
###### for문의 range함수 안에 있는 숫자를 적어서 출력을 원하는 그래프의 갯수를 쉽게 조정 가능

```python
for i in range(4): # 그래프를 생성하고 싶은 CC_NUM 갯수만큼 반복
    ccnum1_name = df_train_ordered["CC_NUM"].value_counts().index[i] # ccnum 추출

    ccnum1_count = df_train_ordered["CC_NUM"].value_counts().values[i] # 해당 ccnum의 데이터 갯수 추출

    df_temp = df_train_ordered[df_train_ordered["CC_NUM"] == ccnum1_name] # 해당 ccnum만 가지는 df 생성

    df_temp.head(10) # 안에는 날짜 순으로 잘 정렬되어 있다

    plt.scatter(list(range(ccnum1_count)), df_temp["AMT"], s=10, c=df_temp["IS_FRAUD"])
    plt.plot(list(range(ccnum1_count)), df_temp["AMT"])
    plt.show()
```
